### 7. Транспонирование матрицы (списка списков)

**Транспонированную матрицу (список списков) присвойте переменной ```list_transpose```**

**Замечание: Можете поискать что такое транспонирование матрицы в интернете**



````python
list_of_lists = [[123, 21, 51], [1, 3, -10]]

list_transpose = [[list_of_lists[j][i] for j in range(len(list_of_lists))] for i in range(len(list_of_lists[0]))]

list_transpose2 = [ list(i) for i in zip(*list_of_lists)]

print(list_transpose)
print(list_transpose2)
````
